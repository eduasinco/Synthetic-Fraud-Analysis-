```python
def fx(x):
    return str(x[1:-1])

def fx2(x):
    if x == "'U'":
        return 7
    else:
        return int(x[1:-1])


df['category'] = df.category.apply(fx)
df['gender'] = df.gender.apply(fx)
df['age'] = df.age.apply(fx2)

le = preprocessing.LabelEncoder()
le.fit(list(set(df['category'])))
df['category'] = le.transform(df['category'])
```
